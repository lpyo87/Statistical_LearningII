# Semana 8 - Clasificación

Se realiza la importación de las librerías a utilizar para la realización de la clasificación del dataset. La librería make_blobs se utilizará para la generación de datos aleatorios con distribución gaussiana.

```python
from sklearn.datasets import make_blobs
import pandas as pd
```

Llamar a la función make_blobs con 500 muestras, 3 centros y un random_state de 10 para asegurar la reproducibilidad. Se le asigna a las variables **x** _y_ **y** respectivamente.

```python
x, y =make_blobs(n_samples = 500, centers = 3, random_state = 10)
```

Se le asignan nombres a las columnas del dataset de muestra generado por medio de función de pandas.

```python
data = pd.DataFrame(x, columns =['a','b'])
```

Se realiza la importación de seaborn para poder graficar la información generada y clasificada.

```python
import seaborn as sns
```

Se realiza la gráfica de dispersión para el dataset y las variables **x** *y* **y**. De esta manera se observa como está agrupando los datos.

```python
sns.scatterplot(data=data, x='a', y='b')
```

Se importa la librería de KMeans para poder realizar la clasificación no supervisada donde se agrupan los objetos de las variables **x** *y* **y**.

```python
from sklearn.cluster import KMeans
```

Se llama a la función KMeans con los parámetros de n_clusters = 2 (indicando número de agrupadores o clusters), n_init = 5 (indica cuantas veces se realizará el cluster con diferentes centroides.) y se asigna a la variable **modelo**. Se obtienen los datos de la variable recién asignada (modelo)

```python
modelo = KMeans(n_clusters = 2, n_init = 5, random_state = 10)
modelo.fit(data)
```

Se obtienen los centros de cluster del modelo por medio del parámetro modelo.cluster_centers_

```python
modelo.cluster_centers_
```

Se visualizan las etiquetas de los modelos generados en el modelo posterior a la clasificación.

```python
modelo.labels_
```

Se obtiene la inercia del modelo el cual es la suma del cuadrado de la distancia de las muestras con respecto al centro de cluster más cercano.

```python
modelo.inertia_
```

Se extrae el parámetro del número de iteraciones realizadas en el modelo.

Se obtiene el parámetro de número de características en el modelo. Se obtienen 2.

```python
modelo.n_features_in_
```

Se realiza una gráfica de dispersión nuevamente la data por medio de la función scatterplot pero para poder identificar las agrupaciones, de las etiquetas, se utilizaron colores naranja y azul para el **1** y **0** respectivamente.

```python
sns.scatterplot(data = data, x='a', y='b', hue = modelo.labels_)
```

Se extraen los centros de los clusters por medio de pandas.

```python
centros = pd.DataFrame(modelo.cluster_centers_, columns = ['x', 'y'])
```

Se llama a la función KMeans con los parámetros de n_clusters = 4 (indicando número de agrupadores o clusters), n_init = 5 (indica cuantas veces se realizará el cluster con diferentes centroides.) y se asigna a la variable **modelo**. Se obtienen los datos de la variable recién asignada (modelo)

```python
modelo = KMeans(n_clusters = 4, n_init = 5, random_state = 10)
modelo.fit(data)
```

Se realiza una gráfica de dispersión nuevamente la data por medio de la función scatterplot pero para poder identificar las agrupaciones, de las etiquetas, se utilizaron establecidos en la paleta Set2 para poder tener una segregación de clusters con respecto a sus centros (marcados en puntos rojos).

```python
sns.scatterplot(data = data, x='a', y='b', hue = modelo.labels_, palette = 'Set2')
sns.scatterplot(data = centros, x='x', y = 'y', color = 'red')
```

Se crean las variables inercias y ks para poder usarlas como depósitos donde ir almacenando los datos por cluster y así determinar el número de clusters adecuado para el análisis de este modelo.

```python
inercias = []
ks = []
```

Con el ciclo For se realiza de manera iterativa el almacenaje de las inercias y número de clusters para posteriormente graficarlo y determinar el núemero óptimo de clusters necesarios para el análisis de los datos.

```python
for k in range(1,5):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data)
    inercias.append(modelo_iterado.inertia_)
    ks.append(k)
```

Se establece una variable con los diccionarios de cluster y número de inercias para determinar su comportamiento. Se asigna a la variable data_codo.

```python
temp = {
    'k':ks,
    'inercia':inercias
}

data_codo = pd.DataFrame(temp)
```

Con la librería seaborn, se grafica de manera lineal los datos de las variables x *y* y que representan el número de clusters y la inercia. 

```python
sns.lineplot(data = data_codo, x='k', y= 'inercia')
```
